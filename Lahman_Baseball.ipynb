```python
import pandas as pd
from sqlalchemy import create_engine

# establish a database connection
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/lahman_baseball")
# Replace `<lahman_baseball_database_name>` with the actual name of your lahman baseball database as it appears in pgadmin

# use the connection to run a query using pandas!
df = pd.read_sql("SELECT * FROM batting;", con=engine)
df.head()
```

# Creating a college players df

```python
Collegeplaying = pd.read_sql("SELECT * from collegeplaying;", con=engine)
Collegeplaying.head()
```

# Extracting Vandy players from collegeplayers df

```python
Vandy_Players = Collegeplaying[Collegeplaying['schoolid'] == 'vandy']
df.head()
```

# creating peoples df

```python
People = pd.read_sql("SELECT * from people;", con=engine)
People.head()
```

# Merging the Vanmdy players df and the peoples df on players id 

```python
Vandy_players = pd.merge(Vandy_Players,People,how="inner")
Vandy_players.head()
```

# Extracting the salaries and creating a df on salaries 

```python
salaries = pd.read_sql("SELECT * from salaries;", con=engine)
print(salaries)
salaries.head()
```

# Merging salaries and Vandy_players to get salaries for vandy players 

```python
Vandy_p_Sal= pd.merge(Vandy_Players,salaries, how="inner", on="playerid")
Vandy_p_Sal.head()
```
